Basic preprocessing (a) check missing & duplicate values

```python
print("Missing values per column :\n",df.isna().sum())
```

```python
print("Duplicate rows:",df.duplicated().sum())
```

```python
num_cols=iris.feature_names   #fill missing numeric with mean
for c in num_cols:
  df[c]=df[c].fillna(df[c].mean())
  print("Missing values in ",c,":",df[c].isna().sum())
```
